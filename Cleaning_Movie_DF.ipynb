```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('movie_data_noindex.csv', sep = ",")
```

```python
df
```

```python
#Create Dummy Variable for Top 5 Composers
top_5 = []
for person in df['composer']:
    if (person == 'John Williams' or person == 'Hans Zimmer' or person == 'James Newton Howard' or person == 'Danny Elfman' or person == 'Alan Silvestri'):
        top_5.append(int(1))
    else:
        top_5.append(int(0))
```

```python
#Adding Dummy Variable to DF
df['top_5'] = top_5

#Changing Column Names
df.columns = ['adj_rank', 'budget', 'composer', 'domestic_rank', 'dom_total_gross', 'movie_title', 'runtime', 'year', 'yearly_rank', 'top_5']
df
```

```python
#Subset of data with top 5 composers (163)
top5_df = df[df['top_5'] == 1]
top5_df
```

```python
df.to_csv('clean_movie_data.csv', sep='\t')
```
